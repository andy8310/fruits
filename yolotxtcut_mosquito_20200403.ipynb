# 使用yolo測試影像切割蚊子

```python
import os
from PIL import Image
import numpy as np
img_path = 'D:/star/test/0.7/GoogleV4/train_image'
save_path = 'D:/star/test/0.7/GoogleV4/train'

all_list_name = os.listdir(img_path)

#讀取txt檔案
for x in range(len(all_list_name)):
    if ".txt" in all_list_name[x]:
       
        txt_filename = all_list_name[x]
        #print(txt_filename)
        file=open(img_path+"/"+txt_filename)
        
        #讀取jpg檔案
        jpg_filename = all_list_name[x].rstrip(".txt")+".jpg"
        #print(jpg_filename)
        im = Image.open(img_path+"/"+jpg_filename)
        # 圖片的宽度和高度
        img_size = im.size
        ###print("圖片寬度和高度分别是{}".format(img_size))
        
        if jpg_filename.split(" ")[0]=="SpeciesmixYS" :
            image_label=[]
            print(jpg_filename)
            for line in file.readlines():                
                curLine=line.strip().split(" ")
                floatLine=list(map(float,curLine))#這裡使用的是map函数直接把数據轉化成为float類型   
                image_label.append(floatLine[1:5])
                first_label=curLine[0]               
                print(first_label)
                
                for ln in range(len(image_label)):
                    # 以label座標位置及長寬截取出圖片(cut image)
                    x = image_label[ln][0]*img_size[0]
                    y = image_label[ln][1]*img_size[1]
                    w = image_label[ln][2]*img_size[0]
                    h = image_label[ln][3]*img_size[1]
                    ###print(x,y,w,h) 
                    region = im.crop((int(x-w/2), int(y-h/2), int(x+w/2), int(y+h/2)))
                    ln = ln+1
                region.save(save_path+'/'+first_label+"_"+jpg_filename.split(".jpg")[0]+"_"+str(ln)+".jpg")
                print(save_path+'/'+first_label+"_"+jpg_filename.split(".jpg")[0]+".jpg")
        else: 
            image_label=[]
            print(jpg_filename)
            for line in file.readlines():                
                curLine=line.strip().split(" ")
                floatLine=list(map(float,curLine))#這裡使用的是map函数直接把数據轉化成为float類型   
                image_label.append(floatLine[1:5])
                first_label=curLine[0]               
                print(first_label)
                
                for ln in range(len(image_label)):
                    # 以label座標位置及長寬截取出圖片(cut image)
                    x = image_label[ln][0]*img_size[0]
                    y = image_label[ln][1]*img_size[1]
                    w = image_label[ln][2]*img_size[0]
                    h = image_label[ln][3]*img_size[1]
                    ###print(x,y,w,h) 
                    region = im.crop((int(x-w/2), int(y-h/2), int(x+w/2), int(y+h/2)))
                    ln = ln+1
                region.save(save_path+'/'+jpg_filename.split(".jpg")[0]+"_"+str(ln)+".jpg")
                print(save_path+'/'+jpg_filename.split(".jpg")[0]+"_"+str(ln)+".jpg")
```
